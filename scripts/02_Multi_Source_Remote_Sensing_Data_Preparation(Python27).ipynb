## üß© Data Process
Before running **03_generate_crater_raster_data (Python27).ipynb**, some data needs to be processed:

- The **Narrow Angle Camera (NAC)** images can be browsed and searched via [QuickMap](https://quickmap.lroc.im-ldi.com/).  
Using its **Search Tool**, users can locate the NAC images corresponding to their regions of interest and obtain the download links. 
Once downloaded, the images can be processed by the **ISIS** ([Github link](https://github.com/DOI-USGS/ISIS3?tab=readme-ov-file)). This processing step converts the raw NAC images into georeferenced TIFF files with spatial information. The resulting georeferenced NAC TIFFs can then be used in subsequent `03_generate_crater_raster_data (Python27).ipynb`.


- The **Clementine UVVIS 5-Band Warped Image** dataset contains five spectral bands.  
  To prepare it for subsequent remote sensing analysis and spatial subsetting, each band was separated into individual single-band images.


- The **LRO DEM**, **Derived Olivine Abundance Map** and **Derived Plagioclase Abundance Map** do not provide full coverage across the study area.  
  Therefore, these datasets were resampled and mosaicked to ensure consistent spatial resolution and continuous coverage across different latitude regions.

Since NAC data can be processed directly by ISIS, the following data processing code primarily focuses on the latter two datasets.


### üåà Band Separation of Multispectral Data

```python
# -*- coding: utf-8 -*-
# For Clementine UVVIS 5-Band Warped Image

# Input raster paths (replace with your local paths)
in_raster = r"your path of UVVIS_Warp_Mosaic_5Bands_200m"
output_dir = r"your output directory of Clementine"
out_folder = os.path.join(output_dir, "clementine_singleband")
if not os.path.exists(out_folder):
    os.makedirs(out_folder)
    
# obtain the bands' number
desc = arcpy.Describe(in_raster)
band_count = desc.bandCount
for i in range(1, band_count + 1):
    band_path = in_raster + "\\Band_" + str(i)
    out_raster = os.path.join(out_folder, "UVVIS_Warp_Mosaic_5Bands_200m_cub_Band_{}.tif".format(i))
    arcpy.management.CopyRaster(band_path, out_raster)
    print("‚úÖ Export Band_{}".format(i))
```

### üó∫Ô∏è Raster Resampling and Mosaicking

&nbsp;&nbsp;&nbsp;&nbsp;For <span style="color:#2A9D8F;">LRO DEM</span> data, 120 m/pixel DEMs were used for regions from 60¬∞ to the poles, and 30 m/pixel DEMs for regions from 75¬∞ to the poles. To ensure consistent spatial resolution and projection accuracy, the 120 m polar DEMs were <span style="color:#E67E22;">resampled </span> to 30 m and then <span style="color:#E67E22;">mosaicked</span> with the corresponding high-latitude 30 m DEMs. This workflow can be implemented in <span style="color:#1E90FF;">ArcGIS Map</span> or automated using <span style="color:#1E90FF;">ArcPy</span>.

```python
# -*- coding: utf-8 -*-
# For LRO DEM

# Input raster paths (replace with your local paths for South/North DEM)
low_latitude_dem = r"your path of LDEM_60S_120M.LBL"   # 120 m DEM
high_latitude_dem = r"your path of LDEM_75S_30M.LBL"   # 30 m DEM

# Output directory and file names
workspace = r"your output directory of LRO DEM"
resampled_dem = os.path.join(workspace, "LDEM_60S_Resampled_30m.tif")
mosaic_output = os.path.join(workspace, "LDEM_60S_30M_merge.tif")

# Create output directory if not exists
if not os.path.exists(workspace):
    os.makedirs(workspace)

# Set environment
arcpy.env.workspace = workspace
arcpy.env.overwriteOutput = True

# ---------------------------------------------
# Step 1. Resample 120 m DEM to 30 m resolution
# ---------------------------------------------
print("üîÑ Resampling 120 m DEM to 30 m resolution ...")
arcpy.management.Resample(
    in_raster=low_latitude_dem,
    out_raster=resampled_dem,
    cell_size="30",                # target resolution
    resampling_type="NEAREST"      # Nearest Neighbor interpolation
)
print("‚úÖ Resampling completed ‚Üí {}".format(resampled_dem))

# ---------------------------------------------
# Step 2. Mosaic resampled and high-latitude DEMs
# ---------------------------------------------
print("üß© Mosaicking resampled and high-latitude DEMs ...")
arcpy.management.MosaicToNewRaster(
    input_rasters=[resampled_dem, high_latitude_dem],
    output_location=workspace,
    raster_dataset_name_with_extension=os.path.basename(mosaic_output),
    pixel_type="16_BIT_SIGNED",
    number_of_bands=1,
    mosaic_method="LAST",                # preserve high-res pixels in overlap
    mosaic_colormap_mode="FIRST"         # use colormap of the first raster
)
print("‚úÖ Mosaic completed ‚Üí {}".format(mosaic_output))
```

---
&nbsp;&nbsp;&nbsp;&nbsp;For the <span style="color:#2A9D8F;">Derived Olivine Abundance Map</span> and <span style="color:#2A9D8F;">Derived Plagioclase Abundance Map</span>, the original resolutions of the North and South Pole rasters are **1002 m/pixel**. To match the mid-latitude map (**59 m/pixel**), the polar rasters need to be <span style="color:#E67E22;">resampled</span>. Before resampling, it was necessary to perform a <span style="color:#E67E22;">projection transformation</span> on the polar datasets, as their native projections differ from that of the mid-latitude map. Specifically, the North and South Pole datasets use the <span style="color:#1E90FF;">Moon_2000_Stereographic_North_Pole</span> and <span style="color:#1E90FF;">Moon_2000_Stereographic_South_Pole</span> projections, respectively.

To ensure spatial consistency across all datasets, the polar rasters were reprojected to match the coordinate system of the mid-latitude map (<span style="color:#1E90FF;">SimpleCylindrical_MOON</span>). This reprojection can be performed in ArcMap by using the *Export Raster Data* tool and specifying the appropriate spatial reference under the *Spatial Reference* option. The resulting files are:
- `gridded_olivine_mosaic_south_pole_moon2000.tif`  
- `gridded_olivine_mosaic_north_pole_moon2000.tif`  
- `gridded_plagioclase_mosaic_south_pole_moon2000.tif`  
- `gridded_plagioclase_mosaic_north_pole_moon2000.tif`

&nbsp;&nbsp;&nbsp;&nbsp;After reprojection,the polar datasets can then be <span style="color:#E67E22;">resampled</span> and <span style="color:#E67E22;">mosaicked</span> with the mid-latitude map. This process can be accomplished using the following <span style="color:#1E90FF;">ArcPy</span> script.


```python
# -*- coding: utf-8 -*-
# For Derived Olivine/Plagioclase Abundance Map and Derived Plagioclase Abundance Map

# Input raster paths with SimpleCylindrical_MOON (replace with your local paths for polar Olivine/Plagioclase Abundance Map)
mid_lat_raster = r"your path of OlivinePercent_50N50S.tif" # or Plagioclase
south_pole_raster = r"your path of gridded_olivine_mosaic_south_pole_moon2000.tif"# or Plagioclase
north_pole_raster = r"your path of gridded_olivine_mosaic_north_pole_moon2000.tif"# or Plagioclase

# Output directory and file names
workspace = r"your output directory of Olivine Abundance Map"# or Plagioclase
south_resampled = os.path.join(workspace, "olivine_south_moon2000_resampled.tif")# or Plagioclase
north_resampled = os.path.join(workspace, "olivine_north_moon2000_resampled.tif")# or Plagioclase
mosaic_output = os.path.join(workspace, "olivine_90N90S.tif")# or Plagioclase

# Create output directory if not exists
if not os.path.exists(workspace):
    os.makedirs(workspace)
    
# Set environment
arcpy.env.workspace = workspace
arcpy.env.overwriteOutput = True

# ---------------------------------------------
# Step 1. Get target cell size from mid-lat raster
# ---------------------------------------------
desc = arcpy.Describe(mid_lat_raster)
target_cell_size = desc.meanCellHeight  # use mid-lat raster resolution

# ---------------------------------------------
# Step 2. Resample polar rasters
# ---------------------------------------------
print("üîÑ Resampling South Pole raster ...")
arcpy.management.Resample(
    in_raster=south_pole_raster,
    out_raster=south_resampled,
    cell_size=target_cell_size,
    resampling_type="NEAREST"  # maintain original pixel values
)
print("‚úÖ Resampling completed ‚Üí {}".format(south_resampled))

print("üîÑ Resampling North Pole raster ...")
arcpy.management.Resample(
    in_raster=north_pole_raster,
    out_raster=north_resampled,
    cell_size=target_cell_size,
    resampling_type="NEAREST"
)
print("‚úÖ Resampling completed ‚Üí {}".format(north_resampled))

# ---------------------------------------------
# Step 3. Define target projection (SimpleCylindrical_MOON / Plate_Carree)
# ---------------------------------------------
sr_string = (
    "PROJCS['SimpleCylindrical_MOON',"
    "GEOGCS['GCS_MOON',"
    "DATUM['D_MOON',SPHEROID['MOON',1737400.0,0.0]],"
    "PRIMEM['Reference_Meridian',0.0],UNIT['Degree',0.0174532925199433]],"
    "PROJECTION['Plate_Carree'],"
    "PARAMETER['false_easting',0.0],"
    "PARAMETER['false_northing',0.0],"
    "PARAMETER['central_meridian',0.0],"
    "UNIT['Meter',1.0]]"
)
sr = arcpy.SpatialReference()
sr.loadFromString(sr_string)

# ---------------------------------------------
# Step 4. Mosaic rasters: South ‚Üí North ‚Üí Mid-latitude
# ---------------------------------------------
print("üß© Mosaicking rasters ...")
arcpy.MosaicToNewRaster_management(
    input_rasters="{};{};{}".format(south_resampled, north_resampled, mid_lat_raster),
    output_location=workspace,
    raster_dataset_name_with_extension=os.path.basename(mosaic_output),
    coordinate_system_for_the_raster=sr,
    pixel_type="32_BIT_FLOAT",
    number_of_bands=1,
    mosaic_method="LAST",
    mosaic_colormap_mode="FIRST"
)

print("‚úÖ Mosaic complete ‚Üí {}".format(mosaic_output))
```
